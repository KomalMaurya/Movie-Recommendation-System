```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data = pd.read_csv('/content/movies.csv')
review_data = pd.read_csv('/content/ratings.csv')
```

```python
movies_data.head()
```

```python
review_data.head()
```

```python
movies_data.shape
```

```python
selected_features = ['genres','title','movieId','rating']
print(selected_features)
```

```python
for feature in selected_features:
  if type(feature)=="str":
    movies_data[feature]=movies_data[feature].fillna('')
for feature in selected_features:
  if type(feature)=="int":
    review_data[feature]=review_data[feature].fillna(0)
```

```python
combined_features = movies_data['genres']+" "+movies_data['title']
```

```python
print(combined_features)
```

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vector = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vector)
```

```python
similarity = cosine_similarity(feature_vector)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
movie_name = input("Enter Movie Name :")
```

```python
list_of_all_titles = movies_data['title'].tolist()
```

```python
print(list_of_all_titles)
```

```python
find_close_movie = difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_movie)
```

```python
close_match = find_close_movie[0]
print(close_match)
```

```python
movies_data.index.name='index'
```

```python
index_of_movie = movies_data[movies_data.title == close_match].index.values[0]
print(index_of_movie)
```

```python
similarity_score = list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sorted_similar_movie = sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movie)
```

```python
print("Movie Recommended for you : \n")

i=1

for movie in sorted_similar_movie:
  index=movie[0]
  title_from_index=movies_data[movies_data.index==index]['title'].values[0]
  if(i<30):
    print(i,'.',title_from_index)
    i+=1
```
